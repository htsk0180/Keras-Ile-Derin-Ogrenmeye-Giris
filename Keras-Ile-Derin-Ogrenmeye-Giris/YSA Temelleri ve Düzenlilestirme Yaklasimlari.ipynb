# 4- Düzenlileştirme Yaklaşımları

* `Yapay Öğrenme Modelinden Bağımsız Olarak Başarımı Artırmanın Yolları` 
## Düzenleme ( Regularization )
Elimizde bir model var. Modelin en yüksek kapasite de çalışmasını istiyoruz. Modele müdahale etmeden max. kapasite de iyi çalışması için yapılması gerekenler:
1.   Veri artırma ( data augmentation )
2.   Sentetik veri üretme ( synhetic data generation )
3.   Gürültü ekleme ( adding noise ) 
4.   Seyreltme ( dropout )
5.   Seyrek-Yoğun-Seyrek-Eğitim ( dense - sparse - dense - training )
6.   Erken durdurma (early stopping )
7.   Transfer öğrenme ( transfer learning )



## Test/Geçerleme/Eğitim Kümesi Seçimi

* Bir ya da birden fazla geçerleme kümesi oluşturmak modelin eğitimi sırasında başarımının nasıl ilerlediğini gözlemlememizi sağlar.
Eğitim kümesi = %60
Geçerleme kümesi = %20 ( burayı %5,%4 vs şeklinde de ayırabiliriz.
Test kümesi = %20
* `Veri setimiz çok büyük değilse;` eğitim kümesini %80 den daha büyük seçmek doğru olmaz.
* Veri setimiz çok çok büyükse eğitim kümesi % 98 seçilebilir.

## Doğruluk Değerlendirme ( Accuracy Evaluation )

* Performans kriteri olarak farklı metrikleri de göz önüne almak gerekir.
* TP,TN,FP,FN değerlerine bakılarak yeni metrikler hesaplanabilir.
1.   `Precision ( hassasiyet ) =` TP / TP + FP
2.   `Accuracy ( doğruluk ) = `TP + TN / TP + TN + FP + FN 
3.   `Recall ( kesinlik ) =` TP / TP + FN

## Veri Artırma
* Verilerin gerçek dünyadaki gibi çeşitli olmasıne bekleriz.
* Veri setinde örnekler artıkça başarımda sonsuza kadar artmayacaktır.
* `Veri artırma - data augmentation işlemi :` Aynı veri üzerinde farklı filtreler uygulayarak, veri boyutu ve rotasyonlarını değiştirme işlemidir.
* `Sentetik veri üretme ( synthetic data generation ) işlemi : ` Elimizde gözlüklü erkek, gözlüksüz erkek ve gözlüksüz kadın verileri olsun. Dolayısıyla gözlüklü kadın verisi yok. Diğer 3 kümedeki veriler kullanılarak gözlüklü kadın verileri üretilebilir. ( zaman ve maaliyet bakımından çok düşük değerler elde ederiz. )

## Aşırı Uydurma ( Overfiting ),  Az Uydurma ( Underfiting ) ve Erken Durdurma ( Early Stopping )

# 5- Aktivasyon Fonksiyonları

* En iyi bilinen aktivasyon fonksiyonu sigmoid olmasına rağmen çeşitki aktivasyon fonksiyonları vardır.
* Aktivasyon fonksiyonu kullanmayan bir sinir ağı sınırlı öğrenme gücüne sahiptir.
* Neden doğrusal olmayan aktivasyon fonksiyonlarını ihtiyacımız var ? Çünkü ; YSA'lar evrensel fonksiyon yakınsayıcıları olarak tasarlanmak istenmektedir. 
* Basamak fn'nin türevi sonsuz değer, doğrusal fn'nin değeri ise sabit bir değer olduğu için bu iki fonksiyon pek fazla aktivasyon fonksiyonu olarak kullanılmaz.
* `En çok kullanılan aktivasyon fonksiyonları:` 
1. Sigmoid fonksiyonu: Türevi sıfıra yakınsadığı noktalarda ( fonksiyonın en yüksek ve en düşük değerler aldığı noktalarda ) öğrenme olmaz. Bu probleme Vanishing Gradient Problem ( gradyanların ölmesi problemi ) denir.
2. Hiperbolik tanjant fonksiyonu : Sigmoid'ten farklı olarak aralığın -1 ve +1 de değişiyor olmasıdır. Türevi artacağı için öğrenme hızını doğrudan artırıyor. Fakat yine uç noktalarda yani türevin sıfıra yakınsadığı noktalarda gradyanların kaybolması mevcuttur.
3. ReLU fonksiyonu :  Sıfırdan sonra + yönde +1 ve sonsuza doğru değer alır. Eksi değerler almadığı ve 0' a yakınsadığı için ReLU'nun işlem yükü çok azdır. Fakat yine gradyanların kaybolması vardır.
4. Leaky ReLU : Gradyon kaybolmasını ortadan kaldırmak için matematiksel bir hile ile sızıntı ReLU ismindeki fonksiyon elde edilir.Eksi değerlerde alabilen -sonsuz tarafında bir fonksiyon yazılır. Ancak bu durum da optimizasyon yapmamız gerekecektir. Çünkü işlem yükü artmıştır.
5. Swish Fonksiyonu : Türevi doğrusal ve sabit değerler alan fonksiyonların öğrenme kapasitesi çok yüksek değildir. Swish fonksiyonu ReLU gibi doğrusal bir türev değerine sahip değildir.


```python
# aktivasyon fonksiyonlarının matematiksel olarak çizdirilmesi:

# gerekli kütüphanelerin yüklenmesi:

import math
import matplotlib.pyplot as plt
import numpy as np

# aktivasyon fonksiyonların tanımlanması:

# Sigmoid:
def sigmoid(x):
  a = []
  for i in x:
    a.append(1/(1+math.exp(-i)))
  return a

# Hiperbolik Tanjant:
def tanh(x, derivative = False):
  if(derivative==True):
    return (1-(x**2))
  return np.tanh(x)

# ReLU: 
def re(x):
  b=[]
  for i in x:
    if i<0:
      b.append(0)
    else:
      b.append(i)
  return b

# Leaky ReLU:
def lr(x):
  b=[]
  for i in x:
    if i < 0:
      b.append(i/10)
    else:
      b.append(i)
  return b

# Swish:
swish = sig*x

# grafiklerin oluşturulması için aralıkların verilmesi:

x = np.arange(-3., 3., 0.1)
sig = sigmoid(x)
tanh = tanh(x)
relu = re(x)
leaky_relu = lr(x)
sw = swish

# fonksiyonların ekrana çizdirilmesi:

line_1, = plt.plot(x,sig, label= 'Sigmoid')
line_2, = plt.plot(x,tanh, label= 'Tanh')
line_3, = plt.plot(x,relu, label= 'ReLU')
line_4, = plt.plot(x,leaky_relu, label= 'Leaky_ReLU')
line_5, = plt.plot(x,sw, label='Swish')

plt.legend(handles=[line_1,line_2,line_3,line_4,line_5])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

`Önemli Notlar: `
* Maaliyet hesabını minimize eden ağırlıkların bulunması için kullanılan optimizasyon yöntemi doğrudan öğrenme yeteneğini ve hızını etki eder.
* Veri çok büyükse girdinin modele parçalı şekilde verilmesi gerekir.( mini batch yöntemi )
* `Transfer Learning :` Çok sık kullanılan ve özelikle çok katmanlı modellerde performansı ciddi şekilde etkileyen düzenleme yöntemidir.
* Daha büyük bir veri setinde eğitilmiş modeli daha küçük bir veri setinde kullanma durumuna : **Fine-Tuning** denir.
* `Çoklu Öğrenme ( Multi-task learning ) : ` Aynı anda farklı görevleri çözebilecek modeller geliştirmektir. Örneğin kadın, erkek ve köpeklerin bulunduğu veri setinde hem kadın / erkek ayrımı yapabilecek hemde insan / köpek ayrımını yapabilecek modeller geliştirmek.
