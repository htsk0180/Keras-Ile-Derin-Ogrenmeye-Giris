# Numpy İle Basit Bir RNN Oluşturmak



```python
import numpy as np
```

* `timesteps` : girişteki zaman adımı sayısı
* `input_features` : girdi nitelik uzayının boyutu
* `output_features` : çıktı nitelik uzayının boyutu 

```python
timesteps = 100
input_features = 32
output_features = 64
```

* Basit olması için girdi verisini rastgele gürültü olarak seçelim.


```python
inputs = np.random.random((timesteps, input_features))
```

* Başlangıç durumu için tüm elamanları sıfırdan oluşan bir vektör oluşturalım.

```python
state_t = np.zeros((output_features,))
```

* Rastgele oluşturulan ağırlık matrisleri.

```python
W = np.random.random((output_features,input_features))
U = np.random.random((output_features,output_features))
b = np.random.random((output_features,))
```

* Girdi ve mevcut duruma göre çıktının oluşturulması.

```python
successive_outputs = []
for input_t in inputs:
  output_t = np.tanh(np.dot(W, input_t) + np.dot(U,state_t) + b)
  successive_outputs.append(output_t)
  state_t = output_features

final_output_sequence = np.concatenate(successive_outputs, axis=0)
```

```python
# IMDB Veri Setini Hazırlama.
from keras.datasets import imdb
import tensorflow as tf
from keras.preprocessing import sequence
```

```python
num_features = 1000
maxlen = 500
bach_size = 32
```

```python
print('load data..')
(input_train,y_train), (input_test, y_test) = imdb.load_data(num_words= num_features)
```

```python
print(len(input_train))
print(len(input_test))
```

```python
print('Pad sequence(sample x train)')
input_train = sequence.pad_sequences(input_train,maxlen=maxlen)
input_test = sequence.pad_sequences(input_test,maxlen=maxlen)
print(input_train)
print(input_test)
```

```python
# embedding ve simplernn katmanların eğitilmesi.
from keras.layers import Dense
from keras import layers
```

```python
# basit rnn ile modelleme.
model = Sequential()
model.add(Embedding(num_features,32))
model.add(SimpleRNN(32))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
# basit bir lstm ile modelleme.
model = Sequential()
model.add(Embedding(num_features,32))
model.add(layers.Bidirectional(layers.LSTM(32)))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
# SimpleRNN ile modelin derlenmesi.
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(input_train,y_train, epochs=10, batch_size= 128, validation_split=0.2)
```

```python
# lstm ile modelin derlenmesi.
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(input_train,y_train, epochs=10, batch_size= 128, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
epochs = range(1, len(acc) + 1)
```

```python
plt.plot(epochs,acc,'m*-', label= 'Eğitim Başarımı')
plt.plot(epochs,val_acc,'g', label= 'Doğrulama Başarımı')
plt.title('Eğitim ve Doğrulama İçin Başarım')
plt.legend()

plt.figure()

plt.plot(epochs,loss,'m*-', label= 'Eğitim Kayıbı')
plt.plot(epochs,val_loss,'g', label= 'Doğrulama Kayıbı')
plt.title('Eğitim ve Doğrulama İçin Kayıp')
plt.legend()

plt.show()
```
